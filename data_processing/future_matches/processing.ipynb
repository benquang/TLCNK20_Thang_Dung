```python
common_columns = match_merge_desc.columns.intersection(new_match.columns)
df2_dropped = new_match.drop(columns=new_match.columns.difference(common_columns))

match_merge_desc = pd.concat([match_merge_desc, df2_dropped], ignore_index=True)
```

```python
#add_new_match = add_new_match.sort_values(by='Match_Date', ascending=False)
match_merge_desc['Match_Date'] = pd.to_datetime(match_merge_desc['Match_Date'])
match_merge_desc = match_merge_desc.sort_values(by='Match_Date', ascending=False)
match_merge_desc.head(15)
```

```python
list_all_teams = match_merge_desc['Home_Team'].unique()
```

```python
for team in list_all_teams:

    this_team = match_merge_desc[(match_merge_desc['Home_Team'] == team) | 
                (match_merge_desc['Away_Team'] == team)]

    for i in range(0, len(this_team)-2):
    
        if(this_team.iloc[i,this_team.columns.get_loc('Home_Team')] == team):

            for j in range(6, 43):

                list_2_elements = []

                for k in range(i+1, i+3):
                    if(this_team.iloc[k,this_team.columns.get_loc('Home_Team')] == team):

                        list_2_elements.append(this_team.iloc[k, j])

                    if(this_team.iloc[k,this_team.columns.get_loc('Away_Team')] == team):

                        list_2_elements.append(this_team.iloc[k, j+41])
            
                #
                match_date = this_team.iloc[i,this_team.columns.get_loc('Match_Date')]
                match_hometeam = this_team.iloc[i,this_team.columns.get_loc('Home_Team')]

                match_merge_desc.loc[(match_merge_desc['Match_Date'] == match_date) 
                                     & (match_merge_desc['Home_Team'] == match_hometeam), 
                                match_merge_desc.columns[j]] = sum(list_2_elements) / 2
            
        if(this_team.iloc[i,this_team.columns.get_loc('Away_Team')] == team):

            for j in range(47, 84):
  
                list_2_elements = []

                for k in range(i+1, i+3):
                    if(this_team.iloc[k,this_team.columns.get_loc('Home_Team')] == team):

                        list_2_elements.append(this_team.iloc[k, j-41])

                    if(this_team.iloc[k,this_team.columns.get_loc('Away_Team')] == team):

                        list_2_elements.append(this_team.iloc[k, j])
            
                #
                match_date = this_team.iloc[i,this_team.columns.get_loc('Match_Date')]
                match_hometeam = this_team.iloc[i,this_team.columns.get_loc('Home_Team')]

                match_merge_desc.loc[(match_merge_desc['Match_Date'] == match_date) 
                                     & (match_merge_desc['Home_Team'] == match_hometeam), 
                                match_merge_desc.columns[j]] = sum(list_2_elements) / 2 

    print(team)  
```

Thêm chỉ số Attack, Defense, Midfield cho các trận mới

```python
matchsquad_players = pd.read_csv("../merge/matchsquad_players_completed.csv")
attack_line = ["LS", "ST", "RS","LW", "LF", "CF", "RF", "RW"]
midfield_line = ["LAM", "CAM", "RAM","LM", "LCM", "CM", "RCM", "RM","LDM", "CDM", "RDM"]
defense_line = ["LB", "LCB", "CB", "RCB", "RB","LWB","RWB"]
new_attrs = pd.DataFrame({'Match_Id': [],
                        'Home_Attack': [],
                        'Home_Midfield': [],
                        'Home_Defense': [],
                        'Away_Attack': [],
                        'Away_Midfield': [],
                        'Away_Defense': []})
atr_match_id = []
atr_home_attack = []
atr_home_midfield = []
atr_home_defense = []
atr_away_attack = []
atr_away_midfield = []
atr_away_defense = []

for i in range(0, len(match_merge_desc)):

    match_id = match_merge_desc.iloc[i, match_merge_desc.columns.get_loc('Match_Id')]
    current_match = matchsquad_players[matchsquad_players['Match_Id'] == match_id]

    home_team = match_merge_desc.iloc[i, match_merge_desc.columns.get_loc('Home_Team')]
    away_team = match_merge_desc.iloc[i, match_merge_desc.columns.get_loc('Away_Team')]
    #

    list_home_attack = []
    list_home_midfield = []
    list_home_defense = []
    list_away_attack = []
    list_away_midfield = []
    list_away_defense = []

    for player in current_match.itertuples(index=False):

        if not np.isnan(player.Sofifa_Id):
            player_positions = player.Player_Position.split(' ')
            overall_rating = player.Player_Overall_Rating
        #home_team
        if (player.Team == home_team) and not np.isnan(player.Sofifa_Id):

            for position in player_positions:
                if (position in attack_line):
                    list_home_attack.append(overall_rating)
                if (position in midfield_line):
                    list_home_midfield.append(overall_rating)
                if (position in defense_line):
                    list_home_defense.append(overall_rating)
        #away_team
        if (player.Team == away_team) and not np.isnan(player.Sofifa_Id):

            for position in player_positions:
                if (position in attack_line):
                    list_away_attack.append(overall_rating)
                if (position in midfield_line):
                    list_away_midfield.append(overall_rating)
                if (position in defense_line):
                    list_away_defense.append(overall_rating)

    atr_match_id.append(match_id)
    atr_home_attack.append(sum(list_home_attack) / len(list_home_attack))
    atr_home_midfield.append(sum(list_home_midfield) / len(list_home_midfield))
    atr_home_defense.append(sum(list_home_defense) / len(list_home_defense))
    atr_away_attack.append(sum(list_away_attack) / len(list_away_attack))
    atr_away_midfield.append(sum(list_away_midfield) / len(list_away_midfield))
    atr_away_defense.append(sum(list_away_defense) / len(list_away_defense))

new_attrs['Match_Id'] = atr_match_id
new_attrs['Home_Attack'] = atr_home_attack
new_attrs['Home_Midfield'] = atr_home_midfield
new_attrs['Home_Defense'] = atr_home_defense
new_attrs['Away_Attack'] = atr_away_attack
new_attrs['Away_Midfield'] = atr_away_midfield
new_attrs['Away_Defense'] = atr_away_defense
match_merge_desc_new = pd.merge(match_merge_desc, new_attrs, on='Match_Id', how='inner')
```
